```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
group = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
import numpy as np
df_trans = group.transpose()

fig, ax = plt.subplots(figsize=(18,8))
x = np.arange(4)
w = 0.07
colors = ['red', 'blue', 'purple', 'grey', 'yellow', 'green', 'pink', 'orangered', 'lightblue', 'violet', 'silver', 'gold']

for i, j in enumerate(df_trans.columns):
    ax.bar(x+(w*i), df_trans[j], width=w, label=j, color=colors[i])

plt.legend(loc='upper right')
plt.xticks(x+(w*6 - (w/2)), labels=df_trans.index)
ax.set_title('Mean Sales Video Games by Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')

plt.show()
```

Jawab: Barplot digunakan untuk memvisualisasikan frekuensi data berdasarkan suatu kategori

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
fig, ax = plt.subplots(figsize=(15,8))
bottom = 0

for i, j in enumerate(df_trans.columns):
    ax.bar(x, df_trans[j], width=0.8, label=j, bottom=bottom, color=colors[i])
    bottom += df_trans[j]

plt.legend(loc='upper right')
plt.xticks(x, labels=df_trans.index)
ax.set_title('Mean Sales Video Games by Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
plt.show()
```

Jawab: Stack Barplot digunakan membandingkan nilai total dari beberapa kategori, berbeda dengan Barplot biasa, Stack Barplot dapat memudahkan kita jika ingin mengetahui nilai total dari sebuah kategori
